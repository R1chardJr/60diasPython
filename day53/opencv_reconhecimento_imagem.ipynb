```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('./imagem.jpg')
if image is None:
    print("Erro: Imagem n√£o encontrada")
else:
    print("Imagem carregada com sucesso")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Imagem em tons de cinza')
plt.axis('off')
plt.show()
```

```python
bordas = cv2.Canny(gray, 100, 200)
plt.imshow(bordas, cmap='gray')
plt.title('Removendo bordas')
plt.axis('off')
plt.show()
```

```python
altura, largura = bordas.shape
print(f'Altura: {altura}, Largura: {largura}')
```

```python
# Criando uma imagem RGBA (com canal alfa) para transpar√™ncia
transparente = np.zeros((altura, largura, 4), dtype=np.uint8)

#para cada pixel que n√£o for borda, deixamos transparente
transparente[ bordas != 0] = [255, 255, 255, 255]  
```

```python
plt.imshow(transparente, cmap='gray')
plt.title("Transparente")
plt.axis("off")
plt.show()
```

```python
cv2.imwrite('imagem_transparente.png', transparente)
```
