```python
import matplotlib.pyplot as plt
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
```

```python
soma_valores = sum(valores)
```

```python
plt.figure(figsize=(8,8))
plt.pie(valores, labels=categorias, autopct = "%1.1f%%")
plt.title("Grafico de Pizza")
plt.legend(categorias, title="Categorias", loc="best")
plt.show()
```

```python
valores_ajustados = [1, 10, 100, 1000]
```

```python
plt.figure(figsize=(8,8))
plt.pie(valores_ajustados, labels=categorias, autopct = "%1.1f%%")
plt.title("Grafico de Pizza")
plt.show()
```

Como podemos ver,est ruim de visualizar, entao para isso 

```python
labels_legenda = []
for i in range(len(categorias)):
    porcentagem = round((valores_ajustados[i] / sum(valores_ajustados)) * 100, 1)
    labels_legenda.append(f"{categorias[i]} - {porcentagem}%")
    
fig, ax = plt.subplots(figsize=(8,8))

patches, texts = ax.pie(
    valores_ajustados,
    labels = None,
)

ax.legend(patches, labels_legenda, title = "Categorias", loc = "best")
plt.title("Grafico de Pizza")
plt.show()
```
